**6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.**

```python
# calculate the median value of the data set.

def calculate_median(m):
  for i in m: # for loop to iterate over column names in 'm'
    if (m[i].dtype==int) or (m[i].dtype==float):# ensuring the dtype sholud be int or float type
      median=m[i].median()  # calculate the median of each column
      print(f"Median of {i} is {median}")
```

```python
# Calling the function calculate_median()
calculate_median(df) # adding my dataset 'df' into the function
```

**10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**

```python
# Defining the function categorize_size and add condition based on total_bedrooms
def categorize_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'Small'
    elif 11 <= total_bedrooms < 1000:
        return 'Medium'
    else:
        return 'Large'

# Creating the new column total_bedroom_size and appling the condition to it.
df['total_bedroom_size'] = df['total_bedrooms'].apply(categorize_size)
df
```
